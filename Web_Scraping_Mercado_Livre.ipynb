##Bibliotecas

```python
#Bibliotecas
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

##Classe (Função)

```python
def mercado_livre(pesquisa):

    lista_mercado_livre = []

    url_base = 'https://lista.mercadolivre.com.br/'
    produto_nome = pesquisa

    print('Link ' + url_base + produto_nome + '\n')

    response = requests.get(url_base + produto_nome)
    site = BeautifulSoup(response.text, 'html.parser')

    produtos = site.findAll('div', attrs={'class':"andes-card andes-card--flat andes-card--default ui-search-result ui-search-result--core andes-card--padding-default"})

    for produto in produtos:
        titulo = produto.find('div',attrs={'class':"ui-search-item__group ui-search-item__group--title"})
        link = produto.find('a',attrs={'class':"ui-search-item__group__element ui-search-link"})
        real = produto.find('span', attrs={'class':"price-tag-fraction"})
        centavo = produto.find('span', attrs={ 'class':"price-tag-cents"})

        parcela = produto.find('span', attrs={'class':"price-tag ui-search-price__part"})
        real_parcela = parcela.find('span', attrs={'class':"price-tag-fraction"})
        centavo_parcela = parcela.find('span', attrs={ 'class':"price-tag-cents"})

        print('Produto: ' + titulo.text)
        print('Link: ' + link['href'])
        if (centavo):
            print(f'Preço: R${real.text},{centavo.text}')
        else:
            print(f'Preço: R${real.text},00')
        if (centavo_parcela):
            print(f'Ou em 12x de R${real_parcela.text},{centavo_parcela.text}')
        else:
            print(f'Ou em 12x de R${real_parcela.text},00')
        print('\n\n')
    print(f'\nTotal de resultados: {len(produtos)} itens')
```

##Execuções

```python
mercado_livre('Copo')
```
