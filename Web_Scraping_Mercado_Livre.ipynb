##Bibliotecas

```python
#Bibliotecas
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

##Classe (Função)

```python
def mercado_livre(pesquisa):

    lista_mercado_livre = []
    tracos = 100

    url_base = 'https://lista.mercadolivre.com.br/'
    produto_nome = pesquisa

    print('Link ' + url_base + produto_nome + '\n')

    response = requests.get(url_base + produto_nome)
    site = BeautifulSoup(response.text, 'html.parser')

    produtos = site.findAll('div', attrs={'class':"andes-card andes-card--flat andes-card--default ui-search-result ui-search-result--core andes-card--padding-default"})


    print('-'*tracos)
    print('FORMATO DE LISTA')
    print('-'*tracos)
    
      
    for produto in produtos:
        titulo = produto.find('div',attrs={'class':"ui-search-item__group ui-search-item__group--title"})
        link = produto.find('a',attrs={'class':"ui-search-item__group__element ui-search-link"})
        
        # Valores
        parcela = produto.find('span', attrs={'class':"price-tag ui-search-price__part"})
        real_parcela = parcela.find('span', attrs={'class':"price-tag-fraction"})
        centavo_parcela = parcela.find('span', attrs={ 'class':"price-tag-cents"})
        
        if(centavo_parcela):
            valor = f'{real_parcela.text}.{centavo_parcela.text}'
            valor = float(valor)
            valor_parcela = round(valor/12, 2)
        else:
            valor = f'{real_parcela.text}'
            valor = float(valor)
            valor_parcela = round(valor/12, 2)
        
        valor = f'R$ {valor}'
        valor_parcela = f'12x de R$ {valor_parcela}'

        valor = valor.replace('.', ',')
        valor_parcela = valor_parcela.replace('.', ',')

        



    # OPÇÕES DE RESULTADO


    # OPÇÃO 1: Resultado no formato de Lista (em tela)

        print('Produto: ' + titulo.text)
        print('Link: ' + link['href'])
        print('Preço: ' + valor)
        print('Parcelamento: ' + valor_parcela)
        print('\n')
    


    # OPÇÃO 2: Resultado no formato de Tabela (em tela)

        lista_mercado_livre.append([titulo.text, valor, valor_parcela, link['href']])
    df_lista_mercado_livre = pd.DataFrame(lista_mercado_livre, columns=['Produto', 'Preço', 'Parcelamento', 'Link'])

    print('-'*tracos)
    print('FORMATO DE TABELA')
    print('-'*tracos)

    display(df_lista_mercado_livre)



    # OPÇÃO 2: Resultado em Excel (em arquivo)

    df_lista_mercado_livre.to_excel(f'Lista Mercado Livre - {pesquisa.title()}.xlsx')

```

##Execuções

```python
mercado_livre('copo')
```
